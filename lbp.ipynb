```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalizar(A):
    A = (A - np.amin(A))/ (np.amax(A) - np.amin(A))
    return A

def distance(A, B):
    soma = 0
    for i in range(len(A)):
        soma = soma + (A[i]-B[i])**2
    return np.sqrt(soma)

def histograma(img):
    qtdeLinhas, qtdeColunas = img.shape
    hist = np.zeros(256)

    for i in range(qtdeLinhas):
        for j in range(qtdeColunas):
            cor = img[i,j]
            hist[cor] = hist[cor] + 1
    
    hist = normalizar(hist)        
            
    return hist        


def lbp(img):
    qtdeLinhas, qtdeColunas = img.shape
    img2 = np.zeros((qtdeLinhas, qtdeColunas), dtype=int)
    
    for i in range(1, qtdeLinhas-1):
        for j in range(1, qtdeColunas-1):
            A = img[i-1, j]
            B = img[i-1, j+1]
            C = img[i, j+1]
            D = img[i+1, j+1]
            E = img[i+1, j]
            F = img[i+1, j-1]
            G = img[i, j-1]
            H = img[i-1, j-1]
            Centro = img[i,j]
            soma = 0
            if(A > Centro):
                soma  = soma + (2**7)
            if(B > Centro):
                soma = soma + (2**6)
            if(C > Centro):
                soma = soma + (2**5)
            if(D > Centro):
                soma  = soma + (2**4)
            if(E > Centro):
                soma = soma + (2**3)
            if(F > Centro):
                soma = soma + (2**2)
            if(G > Centro):
                soma  = soma + (2**1)
            if(H > Centro):
                soma = soma + (2**0) 
            
            img2[i,j]  = soma    
    
    return img2


imgA = cv2.imread('textura1.jpg')
imgA = cv2.cvtColor(imgA, cv2.COLOR_BGR2GRAY)
imgA = lbp(imgA)

imgB = cv2.imread('textura2.jpg')
imgB = cv2.cvtColor(imgB, cv2.COLOR_BGR2GRAY)
imgB = lbp(imgB)

imgC = cv2.imread('textura3.jpg')
imgC = cv2.cvtColor(imgC, cv2.COLOR_BGR2GRAY)
imgC = lbp(imgC)

histA = histograma(imgA)
histB = histograma(imgB)
histC = histograma(imgC)

dAB = distance(histA, histB)
print('a distancia da imagem A para a imagem B é: ', dAB)

dAC = distance(histA, histC)
print('a distancia da imagem A para a imagem C é: ', dAC)

if(dAB < dAC):
    print('a imagem A é mais similar à imagem B')
else:
    print('a imagem A é mais similar à imagem C')


plt.plot(histA)
plt.show()
```
