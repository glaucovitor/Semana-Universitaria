```python
#código abaixo utilizamos para detectar os pontos do contorno de um objeto

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('osso.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#seleciona apenas os pixels dentro do intervalo [250,255]
_, imgBinaria = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY) 


fundo = 255
objeto = 0


qtdeLinhas, qtdeColunas, _ = img.shape

imgContorno =  np.zeros((qtdeLinhas, qtdeColunas), dtype=int)

for i in range(1, qtdeLinhas-1):
    for j in range(1, qtdeColunas-1):
        
        if(imgBinaria[i,j] == objeto):
            A = imgBinaria[i-1,j]
            B = imgBinaria[i+1,j]
            C = imgBinaria[i, j+1]
            D = imgBinaria[i, j-1]
            
            if(A == fundo or B == fundo or C == fundo or D == fundo):
                imgContorno[i,j] = 255
                
plt.gray()
plt.imshow(imgContorno)
```

```python
#codigo abaixo usamos para encontrar o centroide do objeto

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('osso.png')
imgBinaria = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

fundo = 255
objeto = 0

qtdeLinhas, qtdeColunas = imgBinaria.shape

centroideX = 0
centroideY = 0
totalPontos = 0

for i in range(1, qtdeLinhas-1):
    for j in range(1, qtdeColunas-1):
        
        if(imgBinaria[i,j] == objeto):            
            centroideX = centroideX + i
            centroideY = centroideY + j
            totalPontos = totalPontos + 1
            
centroideX = int(centroideX/totalPontos)
centroideY = int(centroideY/totalPontos)

centroide = (centroideY, centroideX)

cv2.circle(img, centroide, 1, (0, 255, 0), 10)
                
plt.gray()
plt.imshow(img)
```

```python
#codigoa abaixo utilizamos para calcular a relação de aspecto de uma imagem binária

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('passaro.png')
imgBinaria = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

x,y,w,h = cv2.boundingRect(imgBinaria)
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

if(h > w):
    aspect_ratio = float(w)/h
else:
    aspect_ratio = float(h)/w

print(aspect_ratio)
                
plt.gray()
plt.imshow(img)
```

```python
#codigo abaixo usamos para desenhar as regiões não-convexas de um objeto

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('osso1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, imgBinaria = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY) 
contours, _ = cv2.findContours(imgBinaria, 
                               cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

cnt = contours[0]

hull = cv2.convexHull(cnt, returnPoints=False)
defects = cv2.convexityDefects(cnt, hull)

for i in range(defects.shape[0]):
    s,e,f,d = defects[i,0]
    start = tuple(cnt[s][0])
    end = tuple(cnt[e][0])
    cv2.line(img,start,end,[0,255,0],2)
    

plt.imshow(img)   
```

```python
#codigo abaixo utilizamos para computar a solidez um objeto

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('osso1.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, imgBinaria = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY) 
contours, _ = cv2.findContours(imgBinaria, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 

cnt = contours[0]

hull = cv2.convexHull(contours[0])
hull_area = cv2.contourArea(hull)
area_objeto = cv2.contourArea(contours[0])
solidez = area_objeto / hull_area

print('area nao convexa: ', hull_area)
print('area do objeto: ', area_objeto)
print('solidez: ', solidez)
```
